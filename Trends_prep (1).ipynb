```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt

base_path = "/kaggle/input/trends-assessment-prediction/"
```

```python
# read non .mat data into dataframes
fnc = pd.read_csv(base_path + "fnc.csv")
loading = pd.read_csv(base_path + "loading.csv")
train_scores = pd.read_csv(base_path + "train_scores.csv")
```

```python
# splitting into train and test
loading_train = loading[loading.Id.isin(train_scores.Id)]
loading_test = loading[~loading.Id.isin(train_scores.Id)]
print(loading_train.shape)
print(loading_test.shape)

fnc_train = fnc[fnc.Id.isin(train_scores.Id)]
fnc_test = fnc[~fnc.Id.isin(train_scores.Id)]
print(fnc_train.shape)
print(fnc_test.shape)
```
